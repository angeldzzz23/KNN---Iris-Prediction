## Split the dataset into testing and training sets with the following parameters: test_size=0.4, random_state=10

## Instantiate a KNN object with K=3, train it on the training set and test it on the testing set. Then, calculate the accuracy of your prediction as you learned in Lab3.

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
import ssl
import requests
from io import StringIO

# I used  ssl because I was getting a weird error when I wasnt using it that required requesting it securely
```

```python
url = 'https://raw.githubusercontent.com/mpourhoma/CS4661/master/iris.csv'
response = requests.get(url, verify=False)
iris_df = pd.read_csv(StringIO(response.text))
```

```python
k = 3
my_knn_for_cshome = KNeighborsClassifier(n_neighbors=k)
```

```python
feature_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
X = iris_df[feature_cols]
y = iris_df['species']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)
```

```python
from sklearn.metrics import accuracy_score
```

```python
my_knn_for_cshome.fit(X_train, y_train)
```

```python
y_predict = my_knn_for_cshome.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_predict)
print(accuracy)
```

for K=1, K=3, K=5, K=11, K=15, K=27, K=43 (you can simply use a “for loop,” and save the final accuracy results in a list). Does the accuracy always get better by increasing the value K?

```python
# Repeading part C 

K_values = [1, 3, 5, 11, 15, 27, 43]
accuracy_results = []
for K in K_values:
    # Initialize the KNN classifier with the current K value
    knn = KNeighborsClassifier(n_neighbors=K)
    
    # Train the model
    knn.fit(X_train, y_train)
    
    # Predict on the test data
    y_pred = knn.predict(X_test)
    
    # Calculate the accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Store the accuracy result for the current K
    accuracy_results.append((K, accuracy))

# Print the accuracy results
for K, acc in accuracy_results:
    print(f'Accuracy for K={K}: {acc:.4f}')
```

# E. 

Now, suppose that we would like to make prediction based on only ONE single feature. To find the best single feature, we have to try every feature individually. In other word, we have to repeat part (c) with K=11, four times (each time using only one of the 4 features), and compute the accuracy each time. Then, check the accuracies. 
Which individual feature provide the best accuracy (the best feature)?  Which one is the second best feature? (Note: you have to train, test, and evaluate your model 4 times, each time on a dataset including only one of the features, and save the final accuracy results in a list).


```python
feature_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

```python
# Assuming 'y' is your target variable, e.g., 'species'
y = iris_df['species']  # Replace 'species' with the actual target column if different.

for feature_col in feature_cols:
    # Select a single feature
    X = iris_df[[feature_col]]  # Keep the double brackets for a DataFrame
    
    # Define K and initialize KNN model
    k = 11
    my_knn_for_cshome = KNeighborsClassifier(n_neighbors=k)
    
    # Split the dataset
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)
    
    # Train the model
    my_knn_for_cshome.fit(X_train, y_train)
    
    # Predict on the test set
    y_predict = my_knn_for_cshome.predict(X_test)
    
    # Calculate and print accuracy
    accuracy = accuracy_score(y_test, y_predict)
    print(f"Feature: {feature_col} -> Accuracy: {accuracy:.4f}")
```

# Answer 
1st: 'petal_width'
2nd: 'petal_length'

# F. 
Now, we want to repeat part (e) (with K=11), this time using two features. you have to train, test, and evaluate your model for 6 different cases: using (1st and 2nd features), (1st and 3rd features), (1st and 4th  features), (2nd  and 3rd  features), (2nd and 4th features), (3rd and 4th  features)!    
Which “feature pair” provides the best accuracy?


```python
feature_cols = [
    ['sepal_length', 'sepal_width'], 
    ['sepal_length', 'petal_length'],
    ['sepal_length', 'petal_width'], 
    ['sepal_width', 'petal_length'], 
    ['sepal_width', 'petal_width'], 
    ['petal_length', 'petal_width']
]

# Assuming 'y' is your target variable, e.g., 'species'
y = iris_df['species']  # Replace 'species' with the actual target column if different.

for feature_col in feature_cols:
    # Select the feature columns from the DataFrame
    X  =  iris_df[feature_col]
    
    # Define K and initialize KNN model
    k = 11
    my_knn_for_cshome = KNeighborsClassifier(n_neighbors=k)
    
    # Split the dataset
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)
    
    # Train the model
    my_knn_for_cshome.fit(X_train, y_train)
    
    # Predict on the test set
    y_predict = my_knn_for_cshome.predict(X_test)
    
    # Calculate and print accuracy
    accuracy = accuracy_score(y_test, y_predict)
    print(f"Features: {feature_col} -> Accuracy: {accuracy:.4f}")
```
